```python
import pandas as pd
import numpy as np
from sklearn.cluster import MeanShift
from sklearn.metrics import pairwise_distances
from haversine import haversine, Unit
```

```python
data = pd.read_csv('checkins.dat', sep='\|')
```

```python
data.info()
```

```python
data.set_axis(['id', 'user_id', 'venue_id', 'latitude', 'longitude', 'created_at'], axis=1, inplace=True)
```

```python
data['latitude'].replace('                   ', np.nan, inplace=True)
data['longitude'].replace('                   ', np.nan, inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
ms = MeanShift(bandwidth=0.1)
```

```python
hundred_rows = data[:100000]
hundred_rows.shape
```

```python
X = hundred_rows.loc[:,['latitude', 'longitude']]
X.shape
```

```python
ms.fit(X)
```

```python
ms.cluster_centers_
```

```python
len(np.unique(ms.labels_))
```

```python
len(ms.labels_)
```

```python
val, counts = np.unique(ms.labels_, return_counts=True)
```

```python
val, counts
```

```python
labels_15 = [x for x in val if counts[x]>15] 
labels_15
```

```python
new_centers_list  = {i:tuple(ms.cluster_centers_[i+1]) for i in labels_15}
new_centers_list 
```

```python
new_centers_list[0]
```

```python
offices = {1:(33.751277, -118.188740), 2:(25.867736, -80.324116), 3:(51.503016, -0.075479), 4:(52.378894, 4.885084), 5:(39.366487, 117.036146), 6:(-33.868457, 151.205134)}
offices
```

```python
offices[1]
```

```python
for office in offices:
  for center in new_centers_list:
    dist = []
    dist.append((office, center, pairwise_distances(office.reshape(1, -1), center.reshape(1, -1))[0].item()))
  dist.sort(key=lambda x: x[2] )
  print(dist)
```
