```python
import pandas as pd
import numpy as np
from math import log, exp, pow, sqrt
from numpy.linalg import norm
```

```python
data = pd.read_csv('data-logistic.csv', header=None)
print(data[:,[1,2]])
```

```python
w=np.array([0,0])
C=0
print(w)
```

```python
for index, row in data.iterrows():
    print(row[0], row[1], row[2])
```

```python
def logistic_regression(data,w,C):
    sum = 0
    for index, row in data.iterrows():
        sum += log(1+exp(-row[0]*(w[0]*row[1]+w[1]*row[2]))) 
    return 1/len(data)*sum + 0.5*C*pow(norm(w),2)
```

```python
logistic_regression(data=data, w=w, C=C)
```

```python
def gradient_descent(w_start, k, data, C, func=None):
    w_curr = w_start
    sum_1 = 0
    sum_2 = 0
    for index, row in data.iterrows():
        sum_1 += row[0]*row[1]*(1 - (1/(1+exp(-row[0]*(w_curr[0]*row[1] + w_curr[1]*row[2])))))
        sum_2 += row[0]*row[2]*(1 - (1/(1+exp(-row[0]*(w_curr[0]*row[1] + w_curr[1]*row[2])))))
    w_1 =  w_curr[0] + k*1/len(data)*sum_1 - k*C*w_curr[0]
    w_2 =  w_curr[1] + k*1/len(data)*sum_2 - k*C*w_curr[1]
    w_curr = np.array([w_1, w_2])
    return w_curr
    
    
```

```python
w_new = gradient_descent(w_start=w, k=0.1, data=data, C=0)
```

```python
print(w_new)
```

```python
logistic_regression(data=data, w=w_new, C=C)
```
