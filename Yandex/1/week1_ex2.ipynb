```python
import pandas as pd
```

```python
data = pd.read_csv('titanic.csv', index_col = 'PassengerId')
data.replace(to_replace='female',value = 0, inplace=True)
data.replace(to_replace='male',value = 1, inplace=True)
data
```

```python
data.shape
#data.dropna(inplace=True)
```

```python
X=data[['Pclass','Fare', 'Age', 'Sex', 'Survived' ]].dropna()
X.isnull()

```

```python
Y = X['Survived']
Y.isnull()
```

```python
X=X[['Pclass','Fare', 'Age', 'Sex' ]]
X
```

```python
from sklearn import tree 
import numpy as np
```

```python
decision_tree = tree.DecisionTreeClassifier(random_state=241)
clf = decision_tree.fit(X, Y)
clf.feature_importances_
```
