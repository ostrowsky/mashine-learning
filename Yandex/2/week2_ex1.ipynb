```python
import pandas as pd
data = pd.read_csv('wine.data')
data
```

```python
y = data.iloc[:, 0]
y
```

```python
X = data.iloc[:, 1:14]
X
```

```python
from sklearn.model_selection import KFold
```

```python
cv = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNC
from sklearn.model_selection import cross_val_score
```

```python
for k in range(1,51):
    estimator = KNC(n_neighbors=k)
    quality = cross_val_score(estimator=estimator, X=X, y=y, cv=cv, scoring="accuracy")
    print(k, quality.mean())
```
